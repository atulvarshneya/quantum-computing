```python
    def initcanvas(self):
        canvas = [[" "]*(self.nq*2+2)]
        for i in range(self.nq*2+2):
            if i%2 == 0:
                canvas[0][i] = "q{0:03d} ".format(i//2)
            else:
                canvas[0][i] = "     "
        canvas[0][self.nq*2] = "creg "
        return canvas

    def get1col(self):
        col = [" "]*(self.nq*2+2)
        for i in range(self.nq*2+2):
            if i%2 == 0:
                col[i] = "-"
            else:
                col[i] = " "
        col[self.nq*2] = "="
        return col
    
    def extend(self, canvas):
        col = get1col()
        canvas.append(col)
        return canvas

    def addX(self, canvas,qbits):
        col = get1col()
        col[qbits[0]*2] = "["
        canvas.append(col)
        col = get1col()
        col[qbits[0]*2] = "X"
        canvas.append(col)
        col = get1col()
        col[qbits[0]*2] = "]"
        canvas.append(col)
        canvas = extend(canvas)
        return canvas

    def addH(self,canvas,qbits):
        col = get1col()
        col[qbits[0]*2] = "["
        canvas.append(col)
        col = get1col()
        col[qbits[0]*2] = "H"
        canvas.append(col)
        col = get1col()
        col[qbits[0]*2] = "]"
        canvas.append(col)
        canvas = extend(canvas)
        return canvas

    def addC(self,canvas,qbits):
        col = get1col()
        col[qbits[0]*2] = "["
        col[qbits[1]*2] = "["
        canvas.append(col)
        col = get1col()
        col[qbits[0]*2] = "."
        col[qbits[1]*2] = "X"
        st = min(qbits[0],qbits[1])*2
        en = max(qbits[0],qbits[1])*2
        for i in range(en-st-1):
            col[i+st+1] = "|"
        canvas.append(col)
        col = get1col()
        col[qbits[0]*2] = "]"
        col[qbits[1]*2] = "]"
        canvas.append(col)
        canvas = extend(canvas)
        return canvas

    def addM(self, canvas,qbits):
        for qb in qbits:
            col = get1col()
            col[qb*2] = "["
            canvas.append(col)
            col = get1col()
            col[qb*2] = "M"
            st = qb*2
            en = self.nq*2
            for i in range(en-st-1):
                col[i+st+1] = "|"
            col[en] = "v"
            canvas.append(col)
            col = get1col()
            col[qb*2] = "]"
            canvas.append(col)
            canvas = extend(canvas)
        return canvas

    def draw(self, canvas):
        nrows = len(canvas)
        ncols = len(canvas[0])
        for i in range(ncols):
            for j in range(nrows):
                print(canvas[j][i], end='')
            print()
```
